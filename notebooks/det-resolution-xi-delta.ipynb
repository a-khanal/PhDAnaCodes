**Detector resolution for Xi reconstruction with FTe**

```python
h1 = ROOT.RDataFrame("T","mc_rec_delta_xi_fall2018_out_fte.root").Histo1D(("mcrec", "mcrec", 25, 1.5, 2.5), "mm_fkpskp")
h1.GetXaxis().SetTitle("MM(eK^{+}K^{+}) [GeV/c^{2}]")
h1.GetYaxis().SetTitle("Counts/40 MeV/c^{2}")

h2 = ROOT.RDataFrame("T","mc_gen_delta_xi_fall2018_out_fte.root").Histo1D(("mcgen", "mcgen", 25, 1.5, 2.5), "mm_fkpskp")
h2.GetXaxis().SetTitle("MM(eK^{+}K^{+}) [GeV/c^{2}]")
h2.GetYaxis().SetTitle("Counts/40 MeV/c^{2}")

h1.GetXaxis().SetTitleSize(0.05)
h1.GetYaxis().SetTitleSize(0.05)
h1.GetXaxis().SetTitleOffset(0.9)
#h1.GetYaxis().SetTitleOffset(0.9)
h2.GetXaxis().SetTitleSize(0.05)
h2.GetYaxis().SetTitleSize(0.05)
h2.GetXaxis().SetTitleOffset(0.9)
#h2.GetYaxis().SetTitleOffset(0.9)

h1.SetLineColor(ROOT.kBlue)
h1.SetMarkerStyle(21)
h1.SetMarkerSize(0.8)
h2.SetLineColor(ROOT.kMagenta)
h2.SetMarkerStyle(21)
h2.SetMarkerSize(0.8)

%jsroot on
ROOT.gStyle.SetOptStat(1111)
ROOT.gStyle.SetOptFit(1111)
c = ROOT.TCanvas()
c.Divide(1,2)
c.cd(1)
h2.Draw()
c.cd(2)
h1.Fit("gaus")
c.Draw()
```

```python
h1 = ROOT.RDataFrame("T","mc_rec_delta_xi_fall2018_out_fte.root").Histo1D(("mcrec", "mcrec", 25, 0.9, 1.4), "mm_fkpskpkm")
h1.GetXaxis().SetTitle("MM(eK^{+}K^{+}K^{-}) [GeV/c^{2}]")
h1.GetYaxis().SetTitle("Counts/10 MeV/c^{2}")

h2 = ROOT.RDataFrame("T","mc_gen_delta_xi_fall2018_out_fte.root").Histo1D(("mcgen", "mcgen", 25, 0.9, 1.4), "mm_fkpskpkm")
h2.GetXaxis().SetTitle("MM(eK^{+}K^{+}K^{-}) [GeV/c^{2}]")
h2.GetYaxis().SetTitle("Counts/10 MeV/c^{2}")

h1.GetXaxis().SetTitleSize(0.05)
h1.GetYaxis().SetTitleSize(0.05)
h1.GetXaxis().SetTitleOffset(0.9)
#h1.GetYaxis().SetTitleOffset(0.9)
h2.GetXaxis().SetTitleSize(0.05)
h2.GetYaxis().SetTitleSize(0.05)
h2.GetXaxis().SetTitleOffset(0.9)
#h2.GetYaxis().SetTitleOffset(0.9)

h1.SetLineColor(ROOT.kBlue)
h1.SetMarkerStyle(21)
h1.SetMarkerSize(0.8)
h2.SetLineColor(ROOT.kMagenta)
h2.SetMarkerStyle(21)
h2.SetMarkerSize(0.8)

fn = ROOT.TF1("fn", "gaus(0)+gaus(3)", 0, 3, 6)
#fn = ROOT.TF1("fn", "gaus(0)+landau(3)", 0, 3, 6)
fn.SetParameter(1, 1.115683)
fn.SetParLimits(1, 1.09, 1.14)
fn.SetParameter(2, 0.06) 
fn.SetParLimits(2, 0.02, 0.09)
fn.SetParameter(4, 1.192642)
fn.SetParLimits(4, 1.17, 1.22)
fn.SetParameter(5, 0.06)
fn.SetParLimits(5, 0.02, 0.09)
%jsroot on
c = ROOT.TCanvas()
c.Divide(1,2)
c.cd(1)
h2.Draw()
c.cd(2)
h1.Fit(fn)
c.Draw()
```
